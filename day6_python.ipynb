#시각화
+ 데이터분석 결과를 수치로 보여주는 것보다는  plot형태로 시각화하면 설득력이 높아진다.
+ 데이터 시각화를 보여주는 전형적인 사례 : 앤스콤 4분할 그래프.
> 앤스콤 4분할 그래프는 영국의 프랭크 앤스콤이 데이터를 시각화하지 않고 수치만 확인했을때 발생할 수 있는 함정을 보여주기 위해 만든 그래프이다.
> 평균,분산,상관관계,회귀선의 통계수치가 같아도 그래프의 형태는 다를수 있다.
> 데이터분석 수치에만 의존하면 잘못된 판단을 할 수 있다는 것을 알려주는 좋은 예시이다. 
+ matplotlib : mathworks에서 개발한 매트랩(MATLAB)과 비슷한 형태. 다양한 데이터를 많은 방법으로 도식화 할 수 있도록 하는 파이썬 라이브러리로리이다. 
+ seaborn : Matplotlib을 기반으로 다양한 색상 테마와 통계용 차트 등의 기능을 추가한 시각화 패키지이다. 
+ 주로 사용되는 plot들 
> 1.line plot : 직선  <br>
> 2.bar plot <br>

# 박스 그래프  : 이산형 변수와 연속형 변수를 함께 사용하는 그래프
+ 수치 데이터를 표현하는 하나의 방식이다. 
- 이산형 변수는 Male, Female과 같이 명확하게 구분되는 값을 의미
- 연속형 변수는 Tip과 같이 명확하게 셀 수 없는 범위의 값을 의미
+ 최소값
+ 제 1사분위 수 (Q1)  : 하위25%
+ 제 2사분위 수 또는 중위수 (Q2) 
+ 제 3사분위 수 (Q3) : 상위25%
+ 최대값

```
boxplot = plt.figure(figsize=(5,3)) 
axes1 = boxplot.add_subplot() 

#성별이 여자인 데이터에서 tip열만 추출하여 리스트에 담음
#성별이 남자인 데이터에서 tip열만 추출하여 리스트에 담음
axes1.boxplot( 
    [tips[tips['sex'] == 'Female']['tip'], 
     tips[tips['sex'] == 'Male']['tip']
    ], 
    labels=['Female', 'Male'])


axes1.set_xlabel('성별') 
axes1.set_ylabel('팁') 
axes1.set_title('성별에 따른 팁')

#최대값, 중간값, 최소값 
a = tips[tips['sex'] == 'Female']['tip']  #여성들의 팁   
b = tips[tips['sex'] == 'Male']['tip']    #남성들의 팁   
print("여자", a.min(), a.max(), a.mean())
print("남자", b.min(), b.max(), b.mean())
#25%아래, 75%이상 
```

> 점의 색상을 다르게 한다. 
- 식사비를 추가하려면?

```
def recode_sex(sex):
    if sex == 'Female':
        return 0 
    else:
        return 1
```

```
 tips['sex'].apply(recode_sex)
```

```
#사용자가 작성한 함수를 한번에 데이터프레임의 각행과 열에 적용하여 실행할 수 있게 해주는 메서드이다. 
tips['sex_color'] = tips['sex'].apply(recode_sex)
```

```
#산점도에 성별추가 
scatter_plot = plt.figure(figsize=(10,5)) 
axes1 = scatter_plot.add_subplot() 

#남자, 여자 다른색깔로 ....
axes1.scatter(x=tips['total_bill'], y=tips['tip'], c=tips["sex_color"] ,   
              s=tips['size']*20 ,alpha=0.7 )    #2,3,4,5,6, )
axes1.set_title('점의 사이즈와 투명도 색깔 추가 ') 
axes1.set_xlabel('Total Bill') 
axes1.set_ylabel('Tip') 
```

```
# seaborn 라이브러리

## 회귀선
+ 산점도가 두 양적 변수들의 직선적 관계를 보여 주고 있다면 그래프 상에 직선을 그어 전체적인 형태를 요약할 수 있다.
+ 두 변수들 중에 한 변수가 나머지 변수를 설명하거나 예측하는데 도움이 되는 경우에만 회귀선으로 두 변수들의 관계를 요약할 수 있다.(주어진 x의 값에 대하여 y의 값을 예측)
+ 설명변수 x가 변할 때 반응변수 y가 얼마나 변하는가를 나타내는 직선이다.
```

```
#regplot 메서드 이용 ...산점도 그래프와 회귀선을 그린다. 
ax = plt.subplots() 
ax = sns.regplot(x='total_bill', y='tip', data=tips) ### hue="sex" 불가 

ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip')
```

```
#산점도 그래프의 크기(scatter_kws)와 모양(markers) 조절하기(hue는 색깔 ) 
#fit_reg 옵션으로 회귀선을 제거 fit_reg=False
scatter = sns.lmplot(x="total_bill", y="tip", data=tips, fit_reg=True, hue="sex",
#                     markers=['o','x'], scatter_kws={"s": 12 , "alpha":0.8}, palette="Set1" )
```

```
#비흡연자가, 흡연자 대비 좀 더 가파른 선형관계를 가지는 것을 볼 수 
scatter = sns.lmplot(x="total_bill", y="tip", data=tips, fit_reg=True, hue="smoker",
                     markers=['o','x'], scatter_kws={"s": 12 , "alpha":0.8}, palette="Set1" )
```

```
#비흡연자가, 흡연자 대비 좀 더 가파른 선형관계를 가지는 것을 볼 수 ..... col 옵션을 추가하여 그래프를 별도로 그려볼 수 있다. col_wrap은 한줄에 칼럼개수 
scatter = sns.lmplot(x="total_bill", y="tip", data=tips, fit_reg=True, hue="smoker", col="day",col_wrap=2,
                     markers=['o','x'], scatter_kws={"s": 12 , "alpha":0.8}, palette="Set1" )
```

```
#jointplot : 산점도와 히스토그램을 한번에 그려주는 메서드이다 숫자형 데이터만 표현 가능
joint = sns.jointplot(x='total_bill', y='tip', data=tips ,kind="scatter", height=7)   #회귀,커널 밀도 맞추기--kde  
#'scatter', 'reg', 'resid', 'kde', or 'hex'"

joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Total Bill과 Tip의 Joint Plot', fontsize=10, y=1.03)
```

```
#jointplot : 산점도와 히스토그램을 한번에 그려주는 메서드이다 숫자형 데이터만 표현 가능
joint = sns.jointplot(x='total_bill', y='tip', data=tips ,kind="reg", height=5 )   #회귀,커널 밀도 맞추기--kde  
#'scatter', 'reg', 'resid', 'kde', or 'hex'"

joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Total Bill과 Tip의 Joint Plot', fontsize=10, y=1.03)
```

```
#산점도 그래프는 점이 겹쳐보일 경우 점을 구분하기 어렵다는 단점이 있다.
# 육각그래프(hexbin) 는 산점도 그래프의 데이터를 구분하기 쉽게해준다.
# 2차원 표면에 육각형으로 데이터를 쌓아 표현하는 그래프이다.
# 특정데이터의 개수가 많아지면 점점 진한색으로 표현된다.
# kind="hex"를 추가한다.   hex 모양의 밀도를 확인
#'scatter', 'reg', 'resid', 'kde', or 'hex'"

hexbin = sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex") #hex 모양의 밀도를 확인

hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```
#산점도 그래프는 점이 겹쳐보일 경우 점을 구분하기 어렵다는 단점이 있다.
# 육각그래프(hexbin) 는 산점도 그래프의 데이터를 구분하기 쉽게해준다.
# 2차원 표면에 육각형으로 데이터를 쌓아 표현하는 그래프이다.
# 특정데이터의 개수가 많아지면 점점 진한색으로 표현된다.
# kind="hex"를 추가한다.   hex 모양의 밀도를 확인
#'scatter', 'reg', 'resid', 'kde', or 'hex'"

hexbin = sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex") #hex 모양의 밀도를 확인

hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```
#kdeplot 메서드로 만든다. 
#shade=True 값을 이용하여 음영을 줄 수 있다

kde, ax = plt.subplots() 
ax = sns.kdeplot(data=tips['total_bill'],
                 data2=tips['tip'], 
                 shade=True)

ax.set_title('Kernel Density Plot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip') 
```

```
#바그래프는 지정한 변수의 평균을 계산하여 그리기 가능 
#다음은 시간에 따라 지불한 비용의 평균을 바 그래프로 그린 예이다. 
ax = plt.subplots() 
ax = sns.barplot(x='time', y='total_bill', data=tips) 


ax.set_title('Bar plot of average total bill for time of day') 
ax.set_xlabel('Time of day') 
ax.set_ylabel('Average total bill')
```

```
#다음은 성별에 따라 지불한 비용의 평균을 바 그래프로 그린 예이다. 
ax = plt.subplots() 
ax = sns.barplot(x='sex', y='total_bill', data=tips) 
ax.set_title('Bar plot of average total bill for sex') 
ax.set_xlabel('sex') 
ax.set_ylabel('Average total bill')
```

```
#박스그래프는 최소값, 1분위수, 중간값, 3분위수, 최대값, 이상치 등 다양한 통계량을 한 번에 표현하는 그래프이다. 
#데이터의 분산이 모호하게 표현된다. 
ax = plt.subplots() 
ax = sns.boxplot(x='time', y='total_bill', data=tips) 
ax.set_title('Boxplot of total bill by time of day') 
ax.set_xlabel('Time of day') 
ax.set_ylabel('Total Bill') 
```

```
# seaborn 라이브러리로 바이올린 그래프 그리기
+ 바이올린 플롯은 데이터의 분포(곡선부분:뚱뚱한부분)와 범위를 한눈에 보기 쉽게 나타내는 그래프이다. 
+ 박스 플롯 (Box plot)과 비슷하지만 더 실제에 가까운 분포를 알 수 있다는 장점
+ 중심선을 따라 대칭인 KDE 플롯이 있다.
+ 가운데 흰색 점은 중앙값(median)을 나타낸다.
+ 바이올린 중앙의 두꺼운 선은 사분위 범위를 나타낸다.
+ 바이올린 플롯은 KDE 플롯과 Box 플롯을 하나로 합쳐놓은 형태
```

```
#박스그래프에서 데이터의 분산이 모호하게 표현된다.
#박스그래프에 커널 밀도를 추정한 바이올린 그래프를 사용한다. 

fig, ax = plt.subplots() 
sns.violinplot(x='time', y='total_bill', data=tips     ) 
ax.set_title('Violin plot of total bill by time of day') 
ax.set_xlabel('Time of day') 
ax.set_ylabel('Total Bill')
```

```
#seaborn 라이브러리 스타일은 darkgrid, whitegrid, dark, white, ticks 로 5가지 종류가있다. 
#whitegrid : 그래프의 배경에 가로줄이 생긴다. 
sns.set_style('whitegrid') 

#색상추가 ... hue 
violin, ax = plt.subplots() 
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=False)  #split값이 True이면 합쳐서본다. 
plt.show()
```
