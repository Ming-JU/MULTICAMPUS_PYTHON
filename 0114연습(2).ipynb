```
#2
def is_palindrome(word):
    #input("문자열을 입력하시오:",word)
    i =0
    j = len(word)-1
    
    
    while i<j:
        if word[i] != word[j]:
            print("회문이아닙니다.")
        i = i+1
        j = j-1
        
    print( "회문입니다.")
```

```
is_palindrome('abba')
```

```
#5
map={}
keypad = [('abc',2),('def',3),('ghi',4),('jkl',5),('mno',6),('pqrs',7),('tuv',8),('wxyz',9)]

for alpha, su in keypad:
  for ch in alpha.upper():
    map[ch] = su


input_str = input("문자열을 입력하시오")

for ch in input_str:
  print(map[ch],end="")
```

```
#9
tel = {}
while True:
  iname = input("(입력모드) 이름을 입력하세요:")
  if not iname:
    break;
  number = input("전화번호를 입력하세요")
  tel[iname] = number

while True:
  sname = input("(검색모드) 이름을 입력하세요:")
  if not sname:
    print("검색이 완료되었습니다.")
    break;
  print(sname,"의 전화번호는",tel[sname],"입니다.")
```
