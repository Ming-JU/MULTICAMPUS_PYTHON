```
#dictionary 형태 (key, value)의 가변인자. 
def myPrint4(*args,**info):
  print(args[0], args[1])
  total = 0
  for su in args:
    total += su
  print(f"합계는 {total}이다.")
  for key, val in info.items():
    print(f"{key}--->{val}")
```

```
def getGrade(score):
  grade = None
  if score>=90:
    grade = "A"
  elif score>=80:
    grade = "B"
  elif score>=70:
    grade = "C"
  elif score>=60:
    grade = "D"
  else:
    grade="F"
  return grade
```
