```
#파이썬의 모든class의 상위(부모)class....object
class Person:
  def __init__(self, name, age, phone):
    #instance변수 
    self.name = name
    self.age = age

    #외부접근불가 변수(정보은닉, 캡슐화)
    self.__phone = phone 
    print("Person생성된다.") 

  #instance메서드 
  def go(self):
    print("간다")

  def pritInfo(self):
    print(f"이름은 {self.name}이고 나이는 {self.age}이다")

  def set_info(self, name2, age2):
    self.name = name2
    self.age = age2

  #setter/getter 외부접근불가 변수 이용방법 : 함수이용 
  def set_Phone(self, phone):
    self.__phone = phone

  def get_Phone(self):
    return self.__phone
```

```
p1 = Person("정다은", 20, "010-1234-5678")
p1.pritInfo()
p1.__dict__

p1.name
p1.age

#정보은닉 : 해당class만 접근 , 외부접근 불가 
#p1.__phone
print(p1.get_Phone())
p1.set_Phone("010-8888-9999")
print(p1.get_Phone())

p1._Person__phone
```

```
#상속 ....기존class를 물려받아서 class정의 ,수정가능, 추가가능 

class Student(Person):
  def __init__(self, name, age, phone, major):
    super().__init__(name, age, phone) 
    self.major = major

  #부모의 메서드 재정의 (덮어쓰기, override)
  def pritInfo(self):
    super().pritInfo()
    print(f"전공은 {self.major} 입니다. ")

  #추가메서드 
  def study(self):
    print(f"{self.name}학생은 공부합니다.")
```

```
class SmartPhone:
  def __init__(self, model):
    print("생성자")
    self.model = model
  def __del__(self):
    print("소멸자")
  def __str__(self):
    print("스마트폰입니다.")
    return self.model 
  def __repr__(self):
    print("스마트폰입니다.__repr__")
    return "문자열__repr__"

  def __eq__(self, other):
    return self.model == other.model

  def __add__(self, other):
    return self.model + other.model 
```

```
class Car(object):
  #class변수
  carCount=0
  def __init__(self, name):
    #instance변수 
    self.name = name
    #자동차가 생성되면 carCount증가
    Car.carCount += 1

  #특수메서드 재정의 
  def __str__(self):
    return f"자동차의 이름은 {self.name}입니다."
  #instance method  
  def carInfo(self):
    print(f"carInfo : 자동차의 이름은 {self.name}입니다.")

  #class메서드
  @classmethod
  def getCount(cls):
    return cls.carCount

  @staticmethod
  def getCount2():
    print("staticmethod이다.")
    return Car.carCount
```

```
c1 = Car("프라이드")
c2 = Car("그랜져")
c3 = Car("그랜져2")
print(Car.carCount , "개가 생성됨")
```

```
class MyIterator:
  def __init__(self, str):
    self.index = 0
    self.str = str

  def __iter__(self):
    return self
  def __next__(self):
    if self.index >= len(self.str):
      raise StopIteration
    s = self.str[self.index]
    self.index += 1
    return s
   
```

```
obj = MyIterator("Hello")

#for in : __iter__(), __next__(), __next__(), __next__(), __next__(), __next__()
for aa in obj:
  print(aa)
```

```
def decorator22(f):
  def wrapper(aa):
    print(f.__name__ + "함수 호출전이다.")
    f(aa)
    print(f.__name__ + "함수 호출후이다.")

  return wrapper
```

```
def hello():
  print("안녕하세요")

def hello2():
  print("안녕하세요2")
```

```
aa = decorator22(hello)
aa("인호")

bb = decorator22(hello2)
bb("민호")
```

```
@decorator22
def hello(name):
  print(f"{name}안녕하세요")

@decorator22
def hello2(name):
  print(f"{name}님 안녕하세요2")
```
