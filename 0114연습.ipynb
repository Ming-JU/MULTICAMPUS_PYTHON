```
#4
def Mean(a):
 return  sum(a)/len(a)

def up(a):
  cnt = 0
  for i in range(len(a)):

    if a[i] > Mean(a):
      cnt += 1
  return cnt

list1=[]
for i in range(5):

  i = int(input("정수를 입력하세요:"))
  list1.append(i)

print(list1)
print("평균=", Mean(list1))
print("평균을 상회하는 숫자의 개수 = ",up(list1))
```
